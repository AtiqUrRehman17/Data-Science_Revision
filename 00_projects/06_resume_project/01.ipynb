```python
import re 
import string
def clear_fun(text):
    text = text.lower()
    text = re.sub('\[.*?\]', ' ', text)
    text = re.sub("\\W"," ",text) 
    text = re.sub('https?://\S+|www\.\S+', ' ', text)
    text = re.sub('<.*?>+', ' ', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), ' ', text)
    text = re.sub('\n', ' ', text)
    text = re.sub('\w*\d\w*', ' ', text) 
   # text = re.sub('\s+', '', text)
    return text
```

```python
df['Resume'][0]
```

```python
# now to apply the above function on the Resume column
df['Resume'] = df['Resume'].apply(clear_fun)
```
