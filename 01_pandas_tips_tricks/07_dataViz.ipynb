# **Data Viz**

```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# lets load the dataset from seaborn
df = sns.load_dataset('titanic')
```

```python
# now plotting
df.plot(kind = 'scatter',x='age',y = 'fare',color = 'red')
```

```python
# here we can use the groupby function for different plot like
df.groupby('sex')['fare'].mean().plot(kind = 'bar',color='red')
# here we can also add titles to the plot using matplotlib
plt.ylabel('avrage fare')
plt.xlabel('Gender')
```

```python
# here how to use seaborn for making plots
# here we will see barplot
sns.barplot(data = df,x='survived',y='fare')
```

```python
sns.barplot(data=df,x='sex',y='fare',hue='class')
```

```python
import plotly.graph_objects as go

import pandas as pd

# Read data from a csv
z_data = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/api_docs/mt_bruno_elevation.csv')

fig = go.Figure(data=[go.Surface(z=z_data.values)])

fig.update_layout(title='Mt Bruno Elevation', autosize=False,
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))

fig.show()
```
