# **Seaborn And Plotting**

```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Important things for plotting accoruding to the grammer of graphics
1. Data (df)
2. type of the plot
3. x - axis
4. y - axis
5. grouping (hue / color) based on the categorical variable
6. size of the plot
7. color plletes
8. style of the plot 
9. how do u interpret

```python
# now importing dataset from seaborn
df = sns.load_dataset('titanic')
```

```python
df.info()
```

```python
# now selecting plot type which is scatter plot
sns.scatterplot(data=df,x = 'age',y = 'fare',hue='sex')
```

```python
# here selecting the figure size of the plot
plt.figure(figsize=(12,8))
sns.scatterplot(data=df,x = 'age', y = 'fare',hue='sex')
```

```python
# now here if i want to change each scatter pint color
pallete = {'male':'blue','female' : 'yellow'}
sns.scatterplot(data=df,x='age',y='fare',hue='sex',palette=pallete)
```

```python
# here how to add xlabels and ylabels and title
pallete = {'male':'blue','female' : 'yellow'}
sns.scatterplot(data=df,x='age',y='fare',hue='sex',palette=pallete)
plt.ylabel('Age')
plt.xlabel('Fare')
plt.title('Sactter Plot For Age Vs Fare')
```

```python
# here is how to save a plot and whenever u save a plot remember to save with resolovtion of 300dpi and also save it into png file
plt.savefig('./outputs/plots/Scatter_plt.png',dpi = 300)
```

```python
# now slelecting style of the plot
# here selecting the figure size of the plot
plt.figure(figsize=(12,8))
sns.scatterplot(data=df,x = 'age', y = 'fare',hue='sex',size='age')
sns.set_theme(style='darkgrid')
```

## **Interpratation**
1. We have age of Passenger at x-axis and fare at y-axis
2. The sacatter plot shows the relation between age and fare at titanic.
3. we further grouped the passenger with thier gender male (green points) and female (orange points).
4. Two passenger have gave the heighest fare and they might have been travlling in the first class.
5. The oldest person in titanic is also male which age is 80 years.

# Box Plot

```python
pal = {'First':'blue','Second':'yellow','Third':'orange'}
sns.boxplot(data= df,x = 'sex',y = 'age',hue='class',palette=pal)
plt.xlabel('Sex')
plt.ylabel('Fare')
plt.title('Box plot for sex and fare')
sns.set_theme(style='darkgrid')
```
