# **Data Inconsistanceies**

```python
# import libraries
import pandas as pd
```

```python
# here we will ganerate a data which have inconsistances
data = {
    'date':['2020-01-01','01/01/2020','2020-01-01','01-01-2020'],
    'country':['USA','U.S.A','America','United States'],
    'name':['attiq','atiq','attiq','atiq'],
    'sale_2020':[100,200,None,200],
    'sale_2021':[None,150,300,150]
}
```

```python
# here to convert the data onto dataframe
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
# standadizing the data format
df['date'] = pd.to_datetime(df['date'],errors='coerce')
df['date'] = df['date'].dt.strftime('%y-%m-%d') # this code will covert the inconsistances into nan then u can relace it.
df.head()
```

```python
# horomonize the name column format
country_mapping = {'USA':'United States','U.S.A':'United States','America':'United States'}
# applying the function on the coulmn
df['country'] = df['country'].replace(country_mapping)
```

```python
df.head()
```

```python
# correcting the typographical mistake in the name column
df['name'] = df['name'].replace({'attiq':'atiq'})
df.head()
```

```python
# remove the duplicate
df.drop_duplicates()
df.head()
```

```python
# resolving the conradicrty data
# lets assume that the 2021 sale are always greater then 2020 sale
# we will remove rows where the condition is not met
df = df.drop(df[df['sale_2021'] <= df['sale_2020']].index)
df.head()
```
