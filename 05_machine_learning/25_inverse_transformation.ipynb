# Inverse Transformation / Inverse Of the Catigorical Variable / Decoding
inverse transfromation mean The decoding of the encoding variables
We can inverse transformation of the following encoders
- `LabelEncoder`
- `OrdinalEncoder`
- `OneHotEncoder`

```python
# importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder,OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
# load the data
df = sns.load_dataset('titanic')
df.head()
```

```python
# here is to impute the misssing values of age deck,emaberk and embarkes_twon
df.isnull().sum().sort_values(ascending=False)
```

```python
# here is impute the missing values
df.drop('deck',axis=1,inplace=True)
df['age'] = df['age'].fillna(df['age'].mean())
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])
df['embark_town'] = df['embark_town'].fillna(df['embark_town'].mode()[0])
```

```python
df.isnull().sum().sort_values(ascending=False)
```

## **Note**
### For Each Column you have to create a separate encoder for each column AND Also for the reverse transformation as well for encoding the use the Fit_transform and For the Inverse Transformation use the Inverse transform.

```python
# here is to use the labelencoder for the encoding
le_sex = LabelEncoder()
le_embarked = LabelEncoder()
le_class = LabelEncoder()
le_who = LabelEncoder()
le_alive = LabelEncoder()
# fitting the encoders
df['sex'] = le_sex.fit_transform(df['sex'])
df['embarked'] = le_embarked.fit_transform(df['embarked'])
df['class'] = le_class.fit_transform(df['class'])
df['who'] = le_who.fit_transform(df['who'])
df['alive'] = le_alive.fit_transform(df['alive'])
```

```python
# here is to see that it is encoded or not
df.head()
```

```python
# Here to Perform the Inverse Trnasformation
df['sex'] = le_sex.inverse_transform(df['sex'])
df['embarked'] = le_embarked.inverse_transform(df['embarked'])
df['class'] = le_class.inverse_transform(df['class'])
df['who'] = le_who.inverse_transform(df['who'])
df['alive'] = le_alive.inverse_transform(df['alive'])
```
