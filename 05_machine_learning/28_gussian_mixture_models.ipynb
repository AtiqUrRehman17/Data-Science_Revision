# Gussian Mixture Models

```python
# importing libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.datasets import load_iris

# load the data
iris = load_iris()
# extract the feature from the data
X = iris.data[:,:2] # here we selected only two feature for simpalcity

# now call the model with componenets/cluster
gmm = GaussianMixture(n_components=3)

# fit the model on x
gmm.fit(X)

# predict the cluster labels
labels = gmm.predict(X)

# here is to find the score of the x using log-likeihood
log_likeihood = gmm.score(X)

# here is to print the likeihood score
print('The score of the likeihood is :',log_likeihood)
# here is to plot the data
plt.scatter(X[:,0],X[:,1],c=labels,cmap='viridis')
plt.xlabel('Sepal_length')
plt.ylabel('Sepal_width')
plt.title('Gussian Mixture Model on Iris Data')
plt.show()
```

## The Parameters For Finding The Clustring making

```python
# importing libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.datasets import load_iris

# load the data
iris = load_iris()
# extract the feature from the data
X = iris.data[:,:2] # here we selected only two feature for simpalcity

# now call the model with componenets/cluster
gmm = GaussianMixture(n_components=3)

# fit the model on x
gmm.fit(X)

# predict the cluster labels
labels = gmm.predict(X)

# here is to find the score of the x using log-likeihood
log_likeihood = gmm.score(X)
# aic and bic
aic = gmm.aic(X)
bic = gmm.bic(X)

# here is to print the likeihood score
print('The score of the likeihood is :',log_likeihood)
print('The AIC :',aic)
print('The BIC :',bic)
# here is to plot the data
plt.scatter(X[:,0],X[:,1],c=labels,cmap='viridis')
plt.xlabel('Sepal_length')
plt.ylabel('Sepal_width')
plt.title('Gussian Mixture Model on Iris Data')
plt.show()
```
