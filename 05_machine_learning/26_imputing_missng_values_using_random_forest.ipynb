# Imputing Missing Values Using Random Forest Alogorithem

```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
from sklearn.preprocessing import LabelEncoder
```

```python
# load the dataset
df = sns.load_dataset('titanic')
df.head()
```

```python
# checking the missing values in the data
df.isnull().sum().sort_values(ascending=False)
```

```python
# here is to remove the missing values and remove the deck column
df.drop('deck',axis=1,inplace=True)
```

```python
# checking the missing values in the data
df.isnull().sum().sort_values(ascending=False)
```

```python
df.columns
```

```python
# now here we will use the labelencoder to encode the categorical variables
# columns to encode
col_to_encode = ['sex','embarked','class','who','embark_town','alive','alone']
# create a dicionary to store labelencoder for each column
label_encoders = {}
# for loop to apply encoder for each column
for col in col_to_encode:
    # create a new encoder for each column
    le = LabelEncoder()
    # fit and transfrom the data an then inverse transfrom
    df[col] = le.fit_transform(df[col])
# check the few row of the data
df.head()
```

```python
# creating two data sets from the age column one cainting missing valeus while other is without missing values
df_with_missing = df[df['age'].isna()]
df_without_missing = df.dropna()
```

```python
# lets see the shape of the data with missing values and without missing valies
print('The  Orignal dataset is :',df.shape)
print('The data set without missing values is :',df_without_missing.shape)
print('The data set with missing values is :',df_with_missing.shape)
```

## Here we will apply the model on the without missing values data to predict the age

```python
from sklearn.model_selection import train_test_split
```

```python
df_with_missing.head()
```

```python
df_without_missing.head()
```

```python
# splitting the data into x and y
X = df_without_missing.drop(['age'],axis=1)
y = df_without_missing['age']

# train test split the data
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)

# call the model
model = RandomForestRegressor(n_estimators=100,random_state=42)
# fitting the model on x and y
model.fit(X_train,y_train)
# prediction
y_pred = model.predict(X_test)
# evaluating the model
print('the MEA is :',mean_squared_error(y_test,y_pred))
print('the MAE is :',mean_absolute_error(y_test,y_pred))
print('the R Squared is :',r2_score(y_test,y_pred))
```
